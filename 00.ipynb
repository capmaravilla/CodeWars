```python
class Student:
    def __init__(self, name, fives, tens, twenties):
        self.name = name
        self.fives = fives
        self.tens = tens
        self.twenties = twenties


david = Student("David", 5, 5, 5)
phil = Student("Phil", 2, 2, 1)
cam = Student("Cameron", 2, 2, 0)
geoff = Student("Geoff", 0, 3, 0)


def total_money(instancia):
    return instancia.fives * 5 + instancia.tens * 10 + instancia.twenties * 20


def most_money(students):
    el = students[0].name
    mon = total_money(students[0])
    count = 1

    for student in students[1:]:
        if total_money(student) == mon:
            count += 1
        if total_money(student) > mon:
            el = student.name
            mon = total_money(student)
    if count == len(students) and len(students) > 1:
        return "All"
    return el


# print(most_money([cam, geoff, phil]))
# print(most_money([cam, geoff]))
# print(most_money([geoff]))
```

```python
def most_money1(students):
    total = []
    for student in students:
        total.append(
            (student.fives * 5) + (student.tens * 10) + (student.twenties * 20)
        )

    if min(total) == max(total) and len(students) > 1:
        return "all"
    else:
        return students[total.index(max(total))].name
```

```python
money = lambda student: 5 * student.fives + 10 * student.tens + 20 * student.twenties


def most_money2(students):
    if len(students) == 1:
        return students[0].name
    D = {money(student): student.name for student in students}
    return "all" if len(D) == 1 else D[max(D)]
```

```python
def fibo1(n):
    if n < 2:
        return n
    return fibo1(n - 1) + fibo1(n - 2)


fibo1(40)
```

```python
def fibo2(n):
    cache = []
    for i in range(n):
        if i < 2:
            cache.append(i)
        else:
            cache.append(cache[-1] + cache[-2])
    return cache


fibo2(41)
```

```python
contador = 0


def actualizar_contador():
    global contador
    contador += 1
    print(contador)


def actualizar_nonlocal():
    contador = 2

    def actualizar():
        global contador
        contador += 1

    actualizar()
    print(contador)


def actualizar_nonlocal2():
    contador = 2

    def actualizar():
        nonlocal contador
        contador += 1

    actualizar()
    print(contador)


actualizar_nonlocal()
print(contador)
```

```python
recursos_ecosistema = {"agua": 1000, "alimento": 800}


def animal_interactua(animal, comida, bebida=0):
    global recursos_ecosistema
    if animal == "carnívoro":
        if recursos_ecosistema["alimento"] >= comida:
            recursos_ecosistema["alimento"] -= comida
            print(f"Un carnívoro ha consumido {comida} de alimento.")
        else:
            print("Recursos insuficientes en el ecosistema")
        print(f"Estado actual del ecosistema: {recursos_ecosistema}")
    elif animal == "herbívoro":
        if (
            recursos_ecosistema["alimento"] >= comida
            and recursos_ecosistema["agua"] >= bebida
        ):
            recursos_ecosistema["alimento"] -= comida
            recursos_ecosistema["agua"] -= bebida
            print(f"Un herbívoro ha consumido {bebida} de agua y {comida} de alimento.")
        else:
            print("Recursos insuficientes en el ecosistema")
        print(f"Estado actual del ecosistema: {recursos_ecosistema}")


def lluvia(agua_lluvia):
    global recursos_ecosistema
    recursos_ecosistema["agua"] += agua_lluvia
    print(f"¡Ha llovido! Se añadieron {agua_lluvia} unidades de agua al ecosistema.")


print(f"Inicio del día en el ecosistema: {recursos_ecosistema}")
animal_interactua("herbívoro", 100, 200)
animal_interactua("carnívoro", 50)
lluvia(200)
print(f"Fin del día en el ecosistema: {recursos_ecosistema}")
```

```python
class Coche:
    def __init__(self, vel_max, consumo, km=0):
        self.vel_max = vel_max
        self.consumo = consumo
        self.km = km
        self.texto = "es un coche"

    def velocidad_maxima(self):
        print(self.vel_max)

    def consumo_medio(self):
        print(self.consumo)

    def especificaciones(self):
        print(
            f"El coche corre a {self.vel_max}, consume {self.consumo} y tiene {self.km} kilometros"
        )


coche1 = Coche(30, 7, 100)
coche2 = Coche(30, 4)

coche1.especificaciones()
coche2.especificaciones()
```
