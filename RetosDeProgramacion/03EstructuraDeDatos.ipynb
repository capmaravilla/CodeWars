```python



class Person:
    def __init__(self, name, age, job="Parado"):
        self.name = name
        self.age = age
        self.job = job

    def hi(self):
        print(f"Hola {self.name} y soy {self.job}!")

    def work(self):
        if self.job:
            print(f"Cosas de {self.job}")
        else:
            print("Estoy en el paro")


class Programmer(Person):
    def __init__(self, name, age):
        super().__init__(name, age, "Programador")

    def work(self):
        print("Picando codigo...")


persona1 = Programmer("David", 48)
persona2 = Person("Nori", 17)

persona2.work()
```
