```python
def list_fibo_recursive(n):

    def num_fibo(n):
        if n < 2:
            return n
        return num_fibo(n - 1) + num_fibo(n - 2)

    cache = []
    for i in range(n):
        cache.append(num_fibo(i))
    return cache


lista_10_fibonacci = list_fibo_recursive(10)
print(lista_10_fibonacci)
```

```python
def list_fibo_normal(n):
    cache = []
    for i in range(n):
        if i < 2:
            cache.append(i)
        else:
            cache.append(cache[-1] + cache[-2])
    return cache


print(list_fibo_normal(10))
```

```python
from functools import lru_cache


# TODO memoizacion y generadores con fibonacci
# Versión optimizada de list_fibo_recursive con memoización:
def list_fibo_recursive(n):
    @lru_cache(maxsize=None)  # Memoización para evitar cálculos redundantes
    def num_fibo(n):
        if n < 2:
            return n
        return num_fibo(n - 1) + num_fibo(n - 2)

    return [num_fibo(i) for i in range(n)]


lista_10_fibonacci = list_fibo_recursive(10)
print(lista_10_fibonacci)


# Versión optimizada de list_fibo_normal con generadores:
def fibo_generator(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b


print(list(fibo_generator(10)))
```
